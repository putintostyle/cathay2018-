```python
import numpy as np
import pandas as pd
import warnings
import math
warnings.filterwarnings("ignore")

from scipy.stats import chi2_contingency as chi2_test
```

```python
path = "/home/math/putintostyle/cathay2018/cathay2018/"
```

```python
raw_info = pd.read_csv(path+"process_data/split_data/train_data_split_occupation.csv").drop(['OCCUPATION'], axis = 1)
raw_info.MINOR_CAT = raw_info.MINOR_CAT.astype(str)
raw_info.head()
```

```python
raw_info = raw_info.drop(['BEHAVIOR_1', 'BEHAVIOR_2', 'BEHAVIOR_3','STATUS1', 'STATUS2', 'STATUS3','STATUS4','INTEREST1', 'INTEREST2', 'INTEREST3', 'INTEREST4',
       'INTEREST5', 'INTEREST6', 'INTEREST7', 'INTEREST8', 'INTEREST9',
       'INTEREST10','BUY_YEAR'], axis = 1)
raw_info.head()
```

```python
raw_column = raw_info.columns.values
raw_column
```

```python
for i in raw_column[2:3]:
    for j in raw_column[3:4]:
        p_value_df =  raw_info[[i,j]]
        table = np.array(p_value_df.values)  ## 叫出兩兩feature的dataframe轉換成array
        
        X = table[:,0]                       ## 叫出第一個大類別
        clean_X = sorted(list(set(X)))       ## 找出裡面的元素
        
        Y = table[:,1]                       ## 叫出第二個大類別
        clean_Y = sorted(list(set(Y)))       ## 找出裡面的元素
        
        ######## 計算列連表 ########
        
        row_table = []
        for mini_label_y in clean_Y:         ## 對於所有Y裡面的元素
            y_container = []                 ## 開一個儲存 for each Y 裡面的元素 -> X裡面的元素 數量的array
                                
            for mini_label_x in clean_X:     ## 對於每個X裡面的類別
                counter = 0                   
                for ele in range(0,len(X)):  ## 對於整個table
                    if (X[ele] == mini_label_x) & (Y[ele] == mini_label_y): ##如Y對，且X也是該類別 
                        counter += 1         ## 計數器 +=1 
                y_container.append(counter)
            row_table.append(y_container)
        row_table = np.array(row_table)
                
contingency_tabel = chi2_test(row_table)

if contingency_tabel[1] < 0.05:
    print('p_valus = ' +  str(list(contingency_tabel)[1]))
    print('independent')
```
