```python
import pandas as pd
import numpy as np
```

```python
path = "/home/math/putintostyle/cathay2018/cathay2018/raw_data/"
```

```python
train_info = pd.read_csv(path+"train_data_concat.csv")
```

```python
train_info.head()
```

# Split Ocupation

```python
ocupation = train_info[['OCCUPATION']]
split_major = []
split_minor = []
for i in ocupation.values:
    if i == ['NANA']:
        print(i)
        split_major.append(np.nan)
        split_minor.append(np.nan)
    else:
        split = list(i[0])
        split_major.append(split[0])
        if len(split) == 3:
            split_minor.append(str(int(float(split[1]+split[2]))))
        elif len(split) == 2:
            split_minor.append(str(int(float(split[1]))))
        else:
            print("RRR")
```

```python
split_occupation_df = pd.DataFrame([split_major,split_minor])
split_occupation_df.head()
```

```python
split_occupation_df = split_occupation_df.T
split_occupation_df.columns = ["MAJOR_CAT", "MINOR_CAT"]
```

# Concat two data 

```python
# train_info = train_info.drop(['OCCUPATION'],axis=1)
train_info = pd.concat([train_info, split_occupation_df], axis=1)
train_info.drop(["OCCUPATION"], axis = 1)
```

```python
train_info.to_csv('/home/math/putintostyle/cathay2018/cathay2018/process_data/split_data/train_data_split_occupation.csv',index=False)
```
